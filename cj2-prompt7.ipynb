```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def compute_exp(x):
    return np.exp(x)
```

```python
y = compute_exp(x)
plt.plot(x, y)
plt.xlabel('Milliseconds')
plt.ylabel('Awesomeness')

plt.savefig('Awesomeness.pdf', bbox_inches='tight', dpi=600)  # Check again what these mean in book
```
