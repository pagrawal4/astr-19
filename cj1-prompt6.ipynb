```python
import math
```

Function that returns sin(x) given a value x

```python
def return_sin(x):
    return np.sin(x)
```

Function that returns cos(x) given a value x

```python
def return_cos(x):
    return np.cos(x)
```

Print out table of sin(x) and cos(x) for 1000 values between 0 and 2*pi

```python
import numpy as np
from tabulate import tabulate

x = np.linspace(0, 2*math.pi, 1000)
y = return_sin(x)
z = return_cos(x)

table_data = [(a, b, c) for a, b, c in zip(x, y, z)]

table_headers = ["x", "sin(x)", "cos(x)"]
python_table = tabulate(table_data, tablefmt="grid", headers=table_headers,
    floatfmt=".3f")

print(python_table)
```

Print out the first 10 values of x, sin(x), and cos(x) in columns

```python
first_10 = table_data[0:10]

print(f"x          sin(x)     cos(x)")

for group in first_10:
    print(f"{group[0]:.4f}     {group[1]:.4f}     {group[2]:.4f}")
```
