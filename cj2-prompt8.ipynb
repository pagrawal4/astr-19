```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def compute_sin(x):
    return np.sin(x)
```

```python
def compute_cos(x):
    return np.cos(x)
```

```python
sinx = compute_sin(x)
cosx = compute_cos(x)

figure, axes = plt.subplots(1, 2)

axes[0].plot(x,sinx)
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].set_title(r'$\sin(x)$')

axes[1].plot(x,cosx)
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].set_title(r'$\cos(x)$')

plt.savefig('Graphs.pdf', bbox_inches='tight', dpi=600)  # Check again what these mean in book
```
